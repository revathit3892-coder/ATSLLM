```python
import string

passwords = input("Enter 5 passwords by comma-seperated:").split(',')

strong_count=0
weak_count=0

for i in passwords:
    #i=i.strip()
    if (len(i) >=3 and 
        any(char.isupper() for char in i) and 
        any(char.islower() for char in i) and 
        any(char.isdigit() for char in i) and 
        any(char in string.punctuation for char in i)):
        strong_count +=1
        print(f"Password {i} is Strong")
    else:
        weak_count +=1
        print(f"Password {i} is Weak")

print("Summary:")
print("Strong passwords:",strong_count)
print("Weak Password:",weak_count)


   
```

```python
user_name = input("Enter user names (comma-seperated):").split(',')

valid_names = 0
invalid_names = 0

for name in user_name:
    if (5<=len(name)<=15 and
        name[0].isalpha() and
        all(c.isalnum() or c=='_' for c in name) and 
        ' ' not in name):
        print(f"Username {name} is Valid")
        valid_names +=1
    else:
        print(f"Username {name} is Invalid")
        invalid_names +=1

print("Summary:")
print("valid",valid_names)
print("Invalid:",invalid_names)
```

```python
import re

usernames = input("Enter usernames (comma-separated): ").split(',')

valid_count = 0
invalid_count = 0

print("\nValidation Results:")
for name in usernames:
    name = name.strip()
    
    if re.fullmatch(r'[A-Za-z][A-Za-z0-9_]{4,14}', name):
        print(f"Username '{name}' is Valid")
        valid_count += 1
    else:
        print(f"Username '{name}' is Invalid")
        invalid_count += 1

print("\nSummary:")
print("Valid:", valid_count)
print("Invalid:", invalid_count)
```
