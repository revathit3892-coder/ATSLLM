```python
# to do list to keep track of tasks:

to_do_list = ['Buy Groceries', 'Pay Bills','Clean the house']

# adding tasks to th list:
to_do_list.append("Schedule Meeting")
to_do_list.append("Go for a Run")

to_do_list.remove("Clean the house")

if 'Pay Bills' in to_do_list:
    print("Dont forget to pay the utility bill")

print("To do list remaining:")
for task in to_do_list:
    print(f"-{task}")
```

```python
# Create a list to store and calculate average grades for students

grades = [85,92,78,90,88]
grades.append(95)

avg_grade = sum(grades)/len(grades)
print("Average Grade:",avg_grade)
print("Highest Grade:",max(grades))
print("Lowest Grade:",min(grades))
```

```python
# Use a list to manage inventory items in a store

inv_items =["apples","bananas","oranges","grapes"]

inv_items.append("Strawberries")
inv_items.remove("bananas")

item="oranges"
if item in inv_items:
    print("In stock")
else:
    print("out of stock")

print("Inventory List:")
for item in inv_items:
    print(f"-{item}")
```

```python
# Use a list to collect and analyze user feedback

feedback = ['Great Service','Very Satisfied','Could be Better','Excellent']

feedback.append('Not happy with the service')

positive_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower() or "satisfied" in comment.lower())
print(positive_count)

print("User feedback:")
for each in feedback:
    print(f"-{each}")
```
